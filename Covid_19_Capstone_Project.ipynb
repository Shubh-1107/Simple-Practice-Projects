```python
import plotly.express as px
```

```python
world = df.groupby(by='Country')[['Confirmed','Active','Deaths','Recovered']].sum().reset_index()
```

```python
figure.show()
```

```python
# Group the data by Date and sum the case types
full_trend = df.groupby('Date')[['Confirmed', 'Deaths', 'Recovered', 'Active']].sum().reset_index()

# Create the line chart using Plotly Express
fig = px.line(full_trend,
              x='Date',
              y=['Confirmed', 'Deaths', 'Recovered', 'Active'],
              title='Global COVID-19 Trends Over Time',
              labels={'value': 'Number of Cases', 'variable': 'Case Type'})

# Chart
fig.update_layout(
    xaxis_title='Date',
    yaxis_title='Number of Cases'
)

# Show the plot
fig.show()
```

```python
# Sort the world DataFrame by 'Recovered' cases in descending order
recovered_by_country = world.sort_values(by='Recovered', ascending=False).reset_index(drop=True)

# Create a bar chart for recovered cases by country
fig_recovered = px.bar(recovered_by_country.head(20), # Display top 20 countries
                       x='Country',
                       y='Recovered',
                       title='Top 20 Countries by Recovered COVID-19 Cases',
                       labels={'Recovered': 'Number of Recovered Cases'},
                       color_discrete_sequence=['red']) # Set a single red color for all bars

# Chart
fig_recovered.update_layout(
    xaxis_title='Country',
    yaxis_title='Number of Recovered Cases',
    xaxis_tickangle=-45 # Rotate country names if they overlap
)

# Show the plot
fig_recovered.show()
```

```python
# Sort the world DataFrame by 'Deaths' in descending order
deaths_by_country = world.sort_values(by='Deaths', ascending=False).reset_index(drop=True)

# Create a bar chart for deaths by country
fig_deaths = px.bar(deaths_by_country,
                    x='Country',
                    y='Deaths',
                    title='Countries - COVID-19 Deaths',
                    labels={'Deaths': 'Number of Deaths'},
                    color_discrete_sequence=['red']) # Set a single red color for all bars

# Chart
fig_deaths.update_layout(
    xaxis_title='Country',
    yaxis_title='Number of Deaths',
    xaxis_tickangle=-45 # Rotate country names if they overlap
)

# Show the plot
fig_deaths.show()
```

```python
# Sort the world DataFrame by 'Deaths' in descending order
deaths_by_country = world.sort_values(by='Deaths', ascending=False).reset_index(drop=True)

# Create a bar chart for deaths by country
fig_deaths = px.bar(deaths_by_country.head(10),#top 10 Countries based on Death
                    x='Country',
                    y='Deaths',
                    title='Top 10 Countries by COVID-19 Deaths',
                    labels={'Deaths': 'Number of Deaths'},
                    color_discrete_sequence=['red']) # Set a single red color for all bars

# Chart
fig_deaths.update_layout(
    xaxis_title='Country',
    yaxis_title='Number of Deaths',
    xaxis_tickangle=-45 # Rotate country names if they overlap
)

# Show the plot
fig_deaths.show()
```

```python
# Sort the world DataFrame by 'Active' cases in descending order
active_by_country = world.sort_values(by='Active', ascending=False).reset_index(drop=True)

# Create a bar chart for active cases by country
fig_active = px.bar(active_by_country.head(20), # Display top 20 countries
                    x='Country',
                    y='Active',
                    title='Top 20 Countries by Active COVID-19 Cases',
                    labels={'Active': 'Number of Active Cases'},
                    color_discrete_sequence=['orange']) # Using orange for active cases

# Chart
fig_active.update_layout(
    xaxis_title='Country',
    yaxis_title='Number of Active Cases',
    xaxis_tickangle=-45 # Rotate country names if they overlap
)

# Show the plot
fig_active.show()
```

```python
# Calculate Recovery Rate for each country

world['Recovery Rate'] = world.apply(lambda row: (row['Recovered'] / row['Confirmed']) * 100 if row['Confirmed'] > 0 else 0, axis=1)

# Sort by Recovery Rate in descending order
recovery_rates_by_country = world.sort_values(by='Recovery Rate', ascending=False).reset_index(drop=True)

# Filter out countries with 0 Confirmed cases that might skew the interpretation of 'rate'
recovery_rates_by_country = recovery_rates_by_country[recovery_rates_by_country['Confirmed'] > 0]

# Select the top N countries for the pie chart (e.g., top 10 or 15)
# A pie chart with too many slices becomes unreadable
top_n_recovered_countries = recovery_rates_by_country.head(10)

# Create the pie chart
fig_pie_recovery = px.pie(top_n_recovered_countries,
                          values='Recovery Rate',
                          names='Country',
                          title='Top 10 Countries by COVID-19 Recovery Rate',
                          hole=0.3, # Creates a donut chart effect
                          color_discrete_sequence=px.colors.qualitative.Pastel) # Use a qualitative color scheme

# Chart
fig_pie_recovery.update_traces(textposition='inside', textinfo='percent+label')
fig_pie_recovery.show()
```

Forecasting for confirmed cases

```python
# renaming confirmed columns to ds and y
confirmed.columns = ['ds','y']
confirmed.head()
```

```python
from prophet import Prophet
m = Prophet()
```

```python
m.fit(confirmed)
```

```python
# Making prediction for next 14 dates
future = m.make_future_dataframe(periods=14)
```

```python
future.tail(15)
```

```python
forecast = m.predict(future)
forecast.tail(15)
```

```python
confirmed
```

```python
forecast_next_14_days = forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(14)
forecast_next_14_days
```

```python
# Cross Validation of train and test
from prophet.diagnostics import cross_validation
df_cv = cross_validation(m, period = '7 days', horizon = '14 days')
df_cv.head()
```

```python
from prophet.diagnostics import performance_metrics
df_p = performance_metrics(df_cv)
df_p.head()
```

```python
from prophet.plot import plot_cross_validation_metric
fig = plot_cross_validation_metric(df_cv, metric='rmse')
```
